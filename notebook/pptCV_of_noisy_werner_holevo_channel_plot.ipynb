# Noisy Werner-Holevo Plot Generator

This notebook generates the plot pertaining to the PPT communication value of the 3-dimensional Werner-Holevo channel being mixed with completely depolarizing channel. That is, we consider $cv^{PPT}(\mathcal{N})$ where
$$ \mathcal{N} = (1-\alpha)\mathcal{N}_{WH,3} + \alpha \Omega_{3} \ , $$
where $\mathcal{N}_{WH,d}$ is the $d$-dimensional Werner-Holevo channel and $\Omega_{d}$ is the completely $d$-dimensional depolarizing channel. <br /> <br />
All data is stored in `*/CVChannel.jl/notebook/plot_data`.

```python
#We use numpy rather than csv because csv reads in csv files as arrays of strings
import matplotlib.pyplot as plt
import numpy as np 
import os
file_base = os.getcwd()+"/plot_data/"
file_to_read = os.path.abspath(file_base+"/noisy-werner-holevo-data.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,6)) #This gets rid of info column
x_axis = data[1:14,0]
#Plot Lines
for i in range(1,6):
    plt.plot(x_axis,data[1:14,i], label = "α = 0."+str((i-1)*5))
plt.xlabel('Channel Parameter, $\lambda$')
plt.ylabel('Non-multiplicativity of $cv^{PPT}(\mathcal{N}^{\otimes 2}) - cv^{PPT}(\mathcal{N})$')
plt.title('$cv^{PPT}$ of Noisy Werner Holevo Channel')
plt.legend()
plt.figure(dpi=1200)
plt.show()
```
