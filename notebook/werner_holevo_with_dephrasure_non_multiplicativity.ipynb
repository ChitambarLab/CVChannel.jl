# Werner-Holevo with Dephrasure CV PPT Non-Multiplicativity

This notebook generates the plot showing the non-multiplicativity of the PPT communication value of the
3-dimensional Werner-Holevo channel run in parallel with the dephrasure channel. That is, we consider $cv^{PPT}(\mathcal{N}_{WH,3,\lambda}\otimes \mathcal{N}_{p,q})$ where $\mathcal{N}_{WH,d,\lambda}$ is the $d$-dimensional Werner-Holevo channel and $\mathcal{N}_{p,q}$ is the dephrasure channel. <br />

All data is stored in `*/cv-channel/notebook/plot_data`.

```python
#We use numpy rather than csv because csv reads in csv files as arrays of strings
import matplotlib.pyplot as plt
import numpy as np 
import os
file_base = os.getcwd()+"/plot_data/"
file_to_read = os.path.abspath(file_base+"/werner-holevo-with-dephrasure-data.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,12)) #This gets rid of info column
x_axis = data[0,1:]
def _label_(j):
    if j < 3:
        return str(0.1*j)
    else:
        return str(0.5+0.1*j)
    
def _ls_(j):
    if j < 3:
        return "-" #(0, (5, 5))
    else:
        return (0, (1, 2))
    
def _lw_(j):
    if j < 3:
        return 2
    else:
        return 4
    
#i: Channel parameter (indexed 0 through 6)
#j: p value (indexed 0 through 5)
for i in range(0,7):
    for j in range(0,6):
        plt.plot(x_axis,data[1+j+6*i,1:], linestyle = _ls_(j), lw=_lw_(j), label ='$p$='+_label_(j))
    plt.title('Non-Multiplicativity of $\mathcal{N}_{WH,3,\lambda} \otimes \mathcal{N}_{p,q}$ for $\lambda$='+str((5*i)/100))
    plt.ylabel('Difference')
    plt.xlabel('$q$ parameter')
    plt.xlim(0, 1)
    plt.ylim(0, 0.45)
    plt.legend()
    #plt.savefig("cv-versus-cvppt-of-WH-with-identity.jpg")
    plt.figure(dpi=1200)
    plt.show()
```
