# Werner-Holevo Non-Multiplicativity Plot

This notebook generates the plot showing the non-multiplicativity of the communication value of the
3-dimensional Werner-Holevo channel. That is, we consider $cv(\mathcal{N}_{WH,3})$ where $\mathcal{N}_{WH,d}$ is the $d$-dimensional Werner-Holevo channel. <br />
Note the data is generated using the $cv^PPT$ optimization program, but all PPT operators are in the relevant 2-fold space are [known to be separable](https://arxiv.org/abs/quant-ph/0010095), which is why this is the true communication value. <br /> <br />
All data is stored in `*/cv-channel/notebook/plot_data`.

```python
#We use numpy rather than csv because csv reads in csv files as arrays of strings
import matplotlib.pyplot as plt
import numpy as np 
import os
file_base = os.getcwd()+"/plot_data/"
file_to_read = os.path.abspath(file_base+"/werner-holevo-non-multiplicativity-data.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,5)) #This gets rid of info column

x_axis = data[:,0]
#Plot Lines
plt.plot(x_axis,data[:,1], label = "$cv(\mathcal{N}_{WH})$")
plt.plot(x_axis,data[:,3], label = "$cv(\mathcal{N}_{WH})^2$")
plt.plot(x_axis,data[:,2], label = "$cv(\mathcal{N}_{WH}^{\otimes 2})$")
plt.plot(x_axis,data[:,4], label = "Activation")
plt.axvline(x=0.3, color='purple', linestyle='--')

plt.xlabel('Channel Parameter, $\lambda$')
plt.title('Non-Multiplicativity of Werner Holevo Channel')
plt.legend()
plt.figure(dpi=1500)
plt.show()
```
