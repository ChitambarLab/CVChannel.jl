# PPT Communication Value Activation with Identity Channel

This notebook shows that while the communication value of the Werner Holevo channel with the identity is stable under tensoring with the identity, the PPT communication value relaxation is not. That is, we look at the difference between $cv(\mathcal{N_{WH,d_{1}}} \otimes \mathrm{id}_{d_{2}})$ and $cv^{PPT}(\mathcal{N_{WH,d_{1}}} \otimes \mathrm{id}_{d_{2}})$ where $\mathcal{N}_{WH,d}$ is the $d$-dimensional Werner-Holevo channel and $\mathrm{id}_{d}$ is the $d$-dimensional identity channel. <br />

We note that we only plot a subset of the data. This is because the optimization programs become unstable as the dimension grows resulting in non-sensical results where the $cv^{PPT}< cv$. 

All data is stored in `*/cv-channel/notebook/plot_data`.

```python
#We use numpy rather than csv because csv reads in csv files as arrays of strings
import matplotlib.pyplot as plt
import numpy as np 
import os
file_base = os.getcwd()+"/plot_data/"
file_to_read = os.path.abspath(file_base+"/cv-ppt-of-werner-holevo-with-identity-data.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,7)) #This gets rid of info column

x_axis = data[1:8,2]
#i: Channel Parameter (1 through 7) 
#j: Werner-Holevo Dimension (1 through 7) 
for i in range(0,3):
    for j in range(0,5):
        plt.plot(x_axis,data[1+(i*49)+(j*7):1+(i*49)+(j*7)+7,6], label ='$d_{1}$='+str(j+2))
    plt.title('$cv_{PPT} - cv$ as a function of $d_{2}$ for $\lambda$='+str(i/10))
    plt.ylabel('Difference')
    plt.xlabel('Werner Holevo Channel Dimension $d_1$')
    plt.xlim(2, 8)
    plt.ylim(0, 4)
    plt.legend()
    plt.figure(dpi=1200)
    plt.show()
```
