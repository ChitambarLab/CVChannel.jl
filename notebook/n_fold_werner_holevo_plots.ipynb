# n-Fold Werner-Holevo Plot Generator

This notebook generates the plots pertaining to the $n$-fold, sometimes high-dimensional, Werner-Holevo channel. That is, it calculates $cv^{PPT}(\mathcal{N}_{WH,d}^{\otimes n})$ for various dimension $d$ and number of copies $n$. We recall
$\mathcal{N}_{WH,d}$ is a mixture of two channels parameterized by channel parameter $\lambda$. <br /> <br />
All data is stored in `*/cv-channel/notebook/plot_data`. The first cell sets up one's ability to retrieve the data, and must be run. <br /> 
All subsections may be run completely independently after that.

```python
#We use numpy rather than csv because csv reads in csv files as arrays of strings
import matplotlib.pyplot as plt
import numpy as np 
import os
file_base = os.getcwd()+"/plot_data/"
```

## Multi-Copy Werner-Holevo PPT Communication Value

```python
file_to_read = os.path.abspath(file_base+"/multi-copy-werner-holevo-script-data.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,12)) #This gets rid of info column
```

```python
x_axis = data[0][1:12]
#Plot Lines
for i in range(1,9):
    plt.plot(x_axis,data[i][1:12], label = "n="+str(i))
plt.ylim([0,65])
plt.xlabel('Channel Parameter, $\lambda$')
plt.ylabel('PPT Communication Value, $cv^{PPT}$')
plt.title('$cv^{PPT}(\mathcal{N}^{\otimes n}_{WH,3})$ channel with parameter $\lambda$')
plt.legend()
plt.figure(dpi=1200)
plt.show()
```

```python
x_axis = data[0][1:12]
#Plot Lines
plt.plot(x_axis,data[11][1:12], label = '$cv(\mathcal{N}^{\otimes 2}) - cv(\mathcal{N})^{2}$')
plt.plot(x_axis,data[12][1:12], label = '$cv^{PPT}(\mathcal{N}^{\otimes 4}) - cv(\mathcal{N}^{\otimes 2})^{2}$')
plt.plot(x_axis,data[13][1:12], label = '$cv^{PPT}(\mathcal{N}^{\otimes 8}) - cv^{PPT}(\mathcal{N}^{\otimes 4})^{2}$')
plt.ylim([0,2])
plt.xlabel('Channel Parameter, $\lambda$')
plt.ylabel('Non-Multiplicativity')
plt.title('Non-Multiplicativity of Werner-Holevo Channel for $d$=3')
plt.legend()
plt.figure(dpi=1200)
plt.show()
```

## Multi-Copy Werner-Holevo Dimension Scaling

```python
file_to_read = os.path.abspath(file_base+"/multi-copy-werner-holevo-dimension-scaling-data.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,38)) #This gets rid of info column
```

```python
x_axis = data[0][1:38]
#Plot lines for =0.0
for i in range(3):
    for j in range(1,5):
        plt.plot(x_axis,data[i*6+j][1:38], label = '$n$='+str(j))
    plt.xlabel('Dimension $d$')
    plt.ylabel('PPT Communication Value $cv^{PPT}$')
    plt.title('Communication Value as a function of dimension for $\lambda$='+str(i*0.25))
    plt.legend()
    plt.figure(dpi=1200)
    plt.show()
```
