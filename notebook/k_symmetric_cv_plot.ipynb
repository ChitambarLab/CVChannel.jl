# Symmetric Extension Communication Value

This notebook generates plots showing an improvement of the $cv^{PPT}$ by instead considering the [$k$-symmetric extension cone](https://arxiv.org/abs/quant-ph/0308032v3) instead of the PPT cone. To do this, we consider a set of channels $\{\mathcal{N}_{\varepsilon}\}_{\varepsilon > 0}$, which are induced by a set of bound bell states parameterized by $\varepsilon > 0$ that are PPT but not separable unless $\varepsilon =  1$. These are presented in [Section 5.1 of this paper]( https://arxiv.org/abs/1004.1655).   <br /> <br />

All data is stored in `*/cv-channel/notebook/plot_data`.

```python
#We use numpy rather than csv because csv reads in csv files as arrays of strings
import matplotlib.pyplot as plt
import numpy as np 
import os
file_base = os.getcwd()+"/plot_data/"
```

## 2-Symmetric Extension Improvement

Here we see an improvemt unsing the 2-symmetric extension

```python
file_to_read = os.path.abspath(file_base+"/two-sym-improvement-bound-bell-data.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,4)) #This gets rid of info column
x_axis = data[1:17,0]
#Plot lines
plt.plot(x_axis, data[1:17,1], label = "$cv^{PPT}(\mathcal{N}_{\epsilon})$")
plt.plot(x_axis,data[1:17,2], label = "$cv^{2-Sym}(\mathcal{N}_{\epsilon})^2$")
#plt.plot(x_axis,data[1:17,3], label = "$Improvement$")

plt.xlabel('Channel Parameter $\epsilon$')
plt.title('Difference Between $cv^{PPT}$ and $cv^{2-Sym}$ as a function of $\epsilon$')
plt.legend()
plt.figure(dpi=1200)
plt.show()
```

## 3-Symmetric Extension Improvement

```python
file_to_read = os.path.abspath(file_base+"/three-sym-improvement-bound-bell-data.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,6)) #This gets rid of info column
x_axis = data[1:8,0]
plt.plot(x_axis, data[1:8,1], label = "$cv^{PPT}(\mathcal{N}_{\epsilon})$")
plt.plot(x_axis,data[1:8,2], label = "$cv^{2-Sym}(\mathcal{N}_{\epsilon})^2$")
plt.plot(x_axis,data[1:8,3], label = "$cv^{3-Sym}(\mathcal{N}_{\epsilon})^2$")

plt.xlabel('Channel Parameter $\epsilon$')
plt.title('Difference Between $cv^{PPT}, cv^{2-Sym}$, and $cv^{3-Sym}$ \n as a function of $\epsilon$')
plt.legend()
plt.figure(dpi=1200)
plt.show()
```
